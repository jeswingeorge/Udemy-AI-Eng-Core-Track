## Types of prompts

You may know this already - but if not, you will get very familiar with it!

Models like GPT have been trained to receive instructions in a particular way.

They expect to receive:

**A system prompt** that tells them what task they are performing and what tone they should use

**A user prompt** -- the conversation starter that they should reply to

````python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
````

````python
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
````

## Messages

The API from OpenAI expects to receive messages in a particular structure.
Many of the other APIs share this structure:

```python
[
    {"role": "system", "content": "system message goes here"},
    {"role": "user", "content": "user message goes here"}
]
```
To give you a preview, the next 2 cells make a rather simple call - we won't stretch the mighty GPT (yet!)

Open AI code

```
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "What is 2 + 2?"}
]

response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
response.choices[0].message.content
```

In OpenAI, the **System Prompt** is treated as just another message in the list. In Gemini, the **System Prompt** is treated as a core part of the "brain" initialization, separate from the conversation itself.

### 1. The Code Translation
Here is how you write that exact OpenAI logic for Gemini:

```python
# 1. Initialize the model with the 'System' instruction
model = genai.GenerativeModel(
    model_name='gemini-2.5-flash',
    system_instruction="You are a helpful assistant"
)

# 2. Make the call with the 'User' content
response = model.generate_content("What is 2 + 2?")

# 3. Print the result
print(response.text)
```

---

### 2. Deep Dive: System vs. User Comparison

To help you follow Ed's course on Windows, keep this mental map of how the two models handle "Roles":

| Concept | OpenAI (Ed's Notebooks) | Gemini (Your Windows Setup) |
| :--- | :--- | :--- |
| **System Prompt** | A dictionary at index `[0]` of the `messages` list. | Defined once inside `genai.GenerativeModel(...)`. |
| **User Prompt** | A dictionary with `role: "user"`. | Passed as a string to `.generate_content()`. |
| **Assistant Role** | Labeled as `"assistant"`. | Labeled as `"model"`. |
| **Updating Persona** | You must re-send the system message in every list. | You define it once; the model "remembers" it for every call. |


### 3. Why Gemini does this
Gemini treats the **System Instruction** as a permanent "identity" for that specific model instance. This is actually more efficient on Windows/VS Code because:
1.  **Cleaner Code**: You don't have to keep appending the system message to a list.
2.  **Safety**: It’s harder for the user to "override" the system instructions in a chat because they are physically separated in the code architecture.

### 4. Windows Tip: "Restart & Run All"
If you change your `system_instruction` in a notebook cell, remember that the `model` variable won't update until you **re-run the cell where you defined the model**. 

> **Pro-tip:** If your Gemini response isn't following your instructions, click the **Restart** button (circular arrow) at the top of your VS Code Notebook to clear the memory and start fresh!

### And now let's build useful messages for GPT-4.1-mini, using a function

```
# See how this function creates exactly the format above - 
# it creates a list of dictionaries, where the first dictionary has the system prompt, and the second dictionary has the user prompt with the website content. 
# This is exactly the format that the model expects when we make a call with both system and user instructions.

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

### Time to bring it together - the API for OpenAI is very simple!
```
# And now: call the OpenAI API. You will get very familiar with this!

def summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```
summarize("https://edwarddonner.com")
```

This is a great moment in the course! Ed is showing you how to **refactor** code—moving from manual lists to a reusable function. This is the "Pro" way to build applications.

**On Windows with Gemini 2.5 Flash, the code is actually **shorter** because we handle the "System Role" during the model setup rather than rebuilding it inside a function every time.**

### 1. The Code Translation

Since you've already defined `model` with your `system_instruction` in the previous step, your translation for this notebook section looks like this:

```python
# Ed's function builds a list of dictionaries.
# For Gemini, we don't need that list structure for the system prompt anymore.
# We just need a function that formats our user prompt.

def prompt_for(website):
    # Simply combine the prefix and the website content
    return user_prompt_prefix + website

# Now, the 'summarize' function
def summarize(url):
    website_content = fetch_website_contents(url)
    
    # We pass the formatted string directly to the model
    # Note: 'model' already has the 'system_instruction' built into it!
    response = model.generate_content(prompt_for(website_content))
    
    return response.text

# Try it out!
print(summarize("https://edwarddonner.com"))
```

---

### 2. Why the "Role" Function is different

Ed's `messages_for` function returns a list because OpenAI requires you to "remind" the model who it is (the system prompt) with every single request.


With **Gemini 2.5 Flash** on your Windows machine:
1. **Model Initialization:** You defined the persona once: `model = genai.GenerativeModel(..., system_instruction=...)`. 
2. **Stateless Calls:** Every time you call `model.generate_content()`, Gemini automatically applies that "System Role" in the background. 
3. **Simpler Logic:** Your `prompt_for` function only needs to worry about the **User** part of the conversation.

---

### 3. Windows-Specific Debugging Tip

* **Variable Scope:** In VS Code Notebooks, if you get a `NameError: name 'user_prompt_prefix' is not defined`, scroll back up to the cell where Ed defined that variable and **run it again**. 
* **Model State:** If you decide to change your "System Prompt" (e.g., from "You are a helpful assistant" to "You are a grumpy pirate"), you **must** re-run the cell where you created the `model` object. Unlike OpenAI's list-based approach, Gemini won't see the change until the object is re-initialized.



````python
# Ed's function builds a list of dictionaries.
# For Gemini, we don't need that list structure for the system prompt anymore.
# We just need a function that formats our user prompt.

# 1. Initialize the model with the 'System' instruction
model = genai.GenerativeModel(
    model_name='gemini-2.5-flash',
    system_instruction=system_prompt
)

def prompt_for(website_content):
    # Simply combine the prefix and the website content
    return user_prompt_prefix + website_content

# Now, the 'summarize' function
def summarize(url):
    website_content = fetch_website_contents(url)
    
    # We pass the formatted string directly to the model
    # Note: 'model' already has the 'system_instruction' built into it!
    response = model.generate_content(prompt_for(website_content))
    return response.text

# Let's try it out on Ed's website
summarize("https://edwarddonner.com")
````

````python
# A function to display this nicely in the output, using markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
````
